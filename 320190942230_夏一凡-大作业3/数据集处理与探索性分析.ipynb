# 1.1.2 数据预处理

```python
import pandas as pd
iris = pd.read_csv('iris.data', sep=',',names=['sepal length', 'sepal width', 'petal length', 'petal width','class'])
iris
```

```python
# 去除缺失值的行
iris = iris.dropna()
#恢复索引
iris.index = range(iris.shape[0])
# 查看数据缺失值情况
iris.isnull().sum()
```

```python
# 对类别标签进行labelencoder处理
from sklearn.preprocessing import StandardScaler, LabelEncoder
le= LabelEncoder()
iris['class'] = le.fit_transform(iris['class'])
iris 
```

```python
# 查看数据的分布情况
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
plt.style.use('Solarize_Light2')
warnings.filterwarnings('ignore')

iris.hist(edgecolor='black', linewidth=1.2)
fig=plt.gcf()
fig.set_size_inches(12,8)
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
sns.violinplot(x='class',y='sepal length',data=iris)
plt.subplot(2,2,2)
sns.violinplot(x='class',y='sepal width',data=iris)
plt.subplot(2,2,3)
sns.violinplot(x='class',y='petal length',data=iris)
plt.subplot(2,2,4)
sns.violinplot(x='class',y='petal width',data=iris)
```

```python
# 将处理好的数据集写入新的文件中
iris.to_csv("preprocessing_iris.csv",index=False)
```

# 1.2.2 数据预处理

```python
import pandas as pd
names = ['Sample code number', 'Clump Thickness', 'Uniformity of Cell Size', 'Uniformity of Cell Shape','Marginal Adhesion', 'Single Epithelial Cell Size', 'Bare Nuclei', 'Bland Chromatin', 'Normal Nucleoli','Mitoses', 'Class']
cancer = pd.read_csv('breast-cancer-wisconsin.data', sep=',', names=names)
cancer
```

可以看出Sample code number列并没有意义，因此在数据预处理过程中我们将该列进行删除

```python
# 去除缺失值的行
cancer = cancer.dropna()
#去除重复值
cancer.drop_duplicates(inplace=True)
#恢复索引
cancer.index = range(cancer.shape[0])
# 查看数据缺失值情况
cancer.isnull().sum()
# 删除Sample列
cancer.drop(['Sample code number'],axis=1,inplace=True)
```

```python
cancer.info()
```

```python
for i, item in enumerate(cancer):
    print('%s:%s'%(item,cancer[item].unique()))
```

可以看出Bare Nuclei列存在一个异常值“？”，这也就是我们在数据集分析前所说的缺失值数量为16的原因

```python
cancer['Bare Nuclei'].value_counts()
```

```python
import numpy as np
import warnings
warnings.filterwarnings('ignore')
cancer['Bare Nuclei'].where(cancer['Bare Nuclei'] != '?',np.nan,inplace=True)
cancer.isnull().sum()
# 于是我们将这些带有异常值的行进行删除
cancer = cancer.dropna()
# 将分类指标2和4映射成0和1
cancer['Class'] =cancer['Class'].map({4:1,2:0})
cancer
```

# 1.2.3 探索性数据分析

```python
# 通过可视化来查看数据的分布情况
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("whitegrid") # 使用whitegrid主题
fig,axes=plt.subplots(nrows=5,ncols=2,figsize=(16,20))
for i, item in enumerate(cancer):
    plt.subplot(5,2,(i+1))
    ax=sns.countplot(item,data = cancer,palette="Pastel1")
    plt.xlabel(str(item),fontsize=14)   
    plt.ylabel('Count',fontsize=14)
    plt.xticks(fontsize=13)
    plt.yticks(fontsize=13)
    i=i+1
    plt.tight_layout()
plt.show()
```
